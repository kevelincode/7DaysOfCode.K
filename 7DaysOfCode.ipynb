```python
#Baixando o arquivo em parquet direto do site da github, para isso foi necessário instalar a biblioteca pyarrow
dados_exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
dados_exemplares
     
```

```python
#Combinando dois dataframes utilizando a coluna comum "codigo de barras"
emprestimos_completo = emprestimos_biblioteca.merge(dados_exemplares)
emprestimos_completo
```

<h1>A quantidade de empréstimos está aumentando ou diminuindo ao decorrer dos últimos anos? </h1>

```python
# Passo 2: Converter a coluna de data para o formato datetime
emprestimos_completo['data_emprestimo'] = pd.to_datetime(emprestimos_completo['data_emprestimo'])

# Passo 3: Extrair o ano da coluna de data
emprestimos_completo['ano'] = emprestimos_completo['data_emprestimo'].dt.year

# Passo 4: Contar a quantidade de ocorrências de empréstimos por ano
emprestimos_por_ano = emprestimos_completo.groupby('ano')['id_emprestimo'].count()

# Exibir os resultados
print(emprestimos_por_ano)
```

<h1>Em quais bibliotecas do sistema estão a maior quantidade de empréstimos?</h1>

<h1>Este é um título grande (h1)</h1>


```python
bibliotecas_sistema = emprestimos_completo.groupby('localizacao')['id_emprestimo'].count()
print(bibliotecas_sistema)
```

```python

# Passo 1: Ordenar as localizações pela quantidade de empréstimos em ordem decrescente
localizacoes_ordenadas = bibliotecas_sistema.sort_values(ascending=False)

# Passo 2: Mostrar as localizações com a maior quantidade de empréstimos (top 10 por exemplo)
top_localizacoes = localizacoes_ordenadas.head(10)
print(top_localizacoes)

# Passo 3: Visualizar os resultados em um gráfico de barras
plt.figure(figsize=(12, 8))
top_localizacoes.plot(kind='bar', color='dodgerblue')
plt.title('Top 10 Localizações com Maior Quantidade de Empréstimos')
plt.xlabel('Localização')
plt.ylabel('Quantidade de Empréstimos')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

```python
# Passo 1: Ordenar as bibliotecas pela quantidade de empréstimos
bibliotecas_ordenadas = bibliotecas_sistema.sort_values(ascending=False)

# Passo 2: Mostrar as bibliotecas com a maior quantidade de empréstimos (top 10 por exemplo)
top_bibliotecas = bibliotecas_ordenadas.head(10)
print(top_bibliotecas)

plt.figure(figsize=(12, 8))
top_bibliotecas.plot(kind='bar', color='skyblue')
plt.title('Top 10 Bibliotecas com Maior Quantidade de Empréstimos')
plt.xlabel('Bibliotecas')
plt.ylabel('Quantidade de Empréstimos')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

<h1> Quais são os temas mais emprestados? E os menos?</h1>

```python
acervo_biblioteca = emprestimos_completo.groupby('colecao')['id_emprestimo'].count()
acervo_biblioteca.head()
```

```python
# Passo 1: Ordenar as bibliotecas pela quantidade de empréstimos
temas_ordenados = acervo_biblioteca.sort_values(ascending=False)

# Passo 2: Mostrar as bibliotecas com a maior quantidade de empréstimos (top 10 por exemplo)
top_temas = temas_ordenados.head(10)
print(top_bibliotecas)

plt.figure(figsize=(12, 8))
top_temas.plot(kind='bar', color='skyblue')
plt.title('Top 10 Temas com Maior Quantidade de Empréstimos')
plt.xlabel('Temas')
plt.ylabel('Quantidade de Temas')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

```python
# Passo 1: Ordenar as bibliotecas pela quantidade de empréstimos
temas_ordenados = acervo_biblioteca.sort_values(ascending=True)

# Passo 2: Mostrar as bibliotecas com a maior quantidade de empréstimos (top 10 por exemplo)
top_temas = temas_ordenados.head(10)
print(top_bibliotecas)

plt.figure(figsize=(12, 8))
top_temas.plot(kind='bar', color='skyblue')
plt.title('Top 10 Temas com Menor Quantidade de Empréstimos')
plt.xlabel('Temas')
plt.ylabel('Quantidade de Temas')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```
